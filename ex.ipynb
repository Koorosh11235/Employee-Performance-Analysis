```python
material_status_dict = {"Single" : 0, "Married":1, "Divorced":2}
df["MaritalStatus"] = df["MaritalStatus"].apply(lambda x : material_status_dict[x])
df["Gender"] = df["Gender"].apply(lambda x : 1 if x == "Male" else 0)
df.head(2)
```

```python
Departments = df["EmpDepartment"].unique()
Job_rols = df["EmpJobRole"].unique()
Bus_tra_fre = df["BusinessTravelFrequency"].unique()
education_background = df["EducationBackground"].unique()
def converting_categorical(category):
    dictionary = {}
    for i, cat in enumerate(category):
        dictionary[cat] = i
    return dictionary
Departments = converting_categorical(Departments)
Job_rols = converting_categorical(Job_rols)
Bus_tra_fre = converting_categorical(Bus_tra_fre)
education_background = converting_categorical(education_background)

df["EducationBackground"] = df["EducationBackground"].apply(lambda x : education_background[x])
df["EmpJobRole"]= df["EmpJobRole"].apply(lambda x : Job_rols[x])
df["BusinessTravelFrequency"] = df["BusinessTravelFrequency"].apply(lambda x : Bus_tra_fre[x])
df["EmpDepartment"] = df["EmpDepartment"].apply(lambda x : Departments[x])
df["Attrition"] = df["Attrition"].apply(lambda x : 1 if x == "Yes" else 0)
df["OverTime"] = df["OverTime"].apply(lambda x : 1 if x == "Yes" else 0)
df.head(7)
```

```python
df_numeric_only = df[['Age','DistanceFromHome','EmpHourlyRate','EmpLastSalaryHikePercent',
                   'TotalWorkExperienceInYears','TrainingTimesLastYear','ExperienceYearsAtThisCompany',
                   'ExperienceYearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']]
df_numeric_only.head(2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20, 18))  # defining canvas size
plotno = 1  # counter

for column in df_numeric_only:  # iteration of columns / accessing the columns from count
    if plotno <= 12:  # set the limit
        plt.subplot(4, 3, plotno)  # plotting 7 graphs (3-rows,3-columns), plotnumber is for count
        sns.boxplot(df_numeric_only[column], orient="h")  # Plotting box plots to detect the outlier
        plt.axvline(df_numeric_only[column].mean(), color="red", linestyle="--", linewidth=2)  # Add mean line
        plt.xlabel(column, fontsize=20)  # assigning name to x-axis and font size is 20
    plotno += 1  # counter increment

plt.tight_layout()
plt.show()  # used to hide the storage location
```
